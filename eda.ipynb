# Week 3 - Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/raw/semi_truck_delivery_data.csv')


sns.scatterplot(x='Distance_Miles', y='Delivery_Time_Hours', data=df)
plt.title('Distance vs Delivery Time')
plt.show()

sns.boxplot(x='Shipping_Type', y='Delivery_Time_Hours', data=df)
plt.title('Shipping Type vs Delivery Time')
plt.xticks(rotation=15)
plt.show()


numeric_df = df.select_dtypes(include=['int64','float64'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```
