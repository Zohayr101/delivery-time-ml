# Week 2 - Data Cleaning & Preprocessing

```python
import pandas as pd

df = pd.read_csv('../data/raw/semi_truck_delivery_data.csv')


df = df.drop_duplicates()


categorical_cols = ['Shipping_Type', 'Traffic_Level', 'Weather_Condition']

for col in categorical_cols:
    df[col] = df[col].astype('category').cat.codes


df['Weight_Category'] = pd.cut(
    df['Cargo_Weight_Lbs'],
    bins=[0,15000,30000,50000],
    labels=['Light Load','Medium Load','Heavy Load']
)

print(df.head())
```
